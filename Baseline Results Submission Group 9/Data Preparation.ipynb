```python
global data_title 
# data_title = pd.DataFrame(columns=['unique_id','asin', 'product_name', 'product_type', 'rating', 'review_title', 'review_text', 'initial_sentiment'])


global uids, asin, product_name, product_type, rating, review_title, review_text
```

```python
global titles 
def reading_file(file_location):
    with open(file_location, 'r') as file:
        global modified_title
        modified_title = ""
        
#         print("Length - ",len(file.readlines()))
#         global data_title
        data = pd.DataFrame(columns=['unique_id','asin', 'product_name', 'product_type', 'rating', 'review_title', 'review_text', 'initial_sentiment'])
        soup = BeautifulSoup(file, 'html.parser')
        global uids, asin, product_name, product_type, rating, review_title, review_text, reviewer_location
        uids = [uid.text.strip() for uid in soup.find_all('unique_id')]
#         print("Length of uids - ", len(uids))
#         print(uids[-3])
        
#         print("Length of titles - ", len(titles))
        
        asin = [asin.text.strip() for asin in soup.find_all('asin')]
#         print("Length of asin - ", len(asin))
        product_name = [pname.text.strip() for pname in soup.find_all('product_name')]
        product_type = [ptype.text.strip() for ptype in soup.find_all('product_type')]
        rating = [rating.text.strip() for rating in soup.find_all('rating')]
        
#         print("Length of rating - ", len(rating))
        review_title = [rtitle.text.strip() for rtitle in soup.find_all('title')]
        review_text = [rtext.text.strip() for rtext in soup.find_all('review_text')]

        # loop through list and modify each item, then append to DataFrame
        for i in range(len(uids)):
            modified_title = review_title[i].strip()  # modify item here
            modified_uid = uids[i].strip().split(':')[0]
    
            if float(rating[i])>=4.0:
                isentiment = "positive"
            elif float(rating[i])<=2.0:
                isentiment = "negative"
            else:
                isentiment = "neutral"
            data = data.append({'unique_id':modified_uid, 'asin':asin[i], 'product_name':product_name[i], 'product_type':product_type[i], 'rating':rating[i], 'review_title': modified_title, 'review_text':review_text[i], 'initial_sentiment':isentiment}, ignore_index=True)
        return data
        
```

```python
file_location = f"{location}\{'books/positive.review'}"
books_positive = reading_file(file_location)

file_location = f"{location}\{'books/negative.review'}"
books_negative = reading_file(file_location)

file_location = f"{location}\{'dvd/positive.review'}"
dvd_positive = reading_file(file_location)

file_location = f"{location}\{'dvd/negative.review'}"
dvd_negative = reading_file(file_location)

file_location = f"{location}\{'dvd/unlabeled.review'}"
dvd_unlabeled = reading_file(file_location)

file_location = f"{location}\{'electronics/positive.review'}"
electronics_positive = reading_file(file_location)

file_location = f"{location}\{'electronics/negative.review'}"
electronics_negative = reading_file(file_location)

file_location = f"{location}\{'electronics/unlabeled.review'}"
electronics_unlabeled = reading_file(file_location)

file_location = f"{location}\{'kitchen_&_housewares/positive.review'}"
kitchen_house_positive = reading_file(file_location)

file_location = f"{location}\{'kitchen_&_housewares/negative.review'}"
kitchen_house_negative = reading_file(file_location)

file_location = f"{location}\{'kitchen_&_housewares/unlabeled.review'}"
kitchen_house_unlabeled = reading_file(file_location)


```
