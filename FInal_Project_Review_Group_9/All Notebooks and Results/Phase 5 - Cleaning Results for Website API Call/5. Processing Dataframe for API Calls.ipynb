```python
# Remove unnecessary
df.drop(['Unnamed: 0'], axis=1, inplace = True)
df
```

```python
# Converting 'aspect_ratings' column from type string to dictionary 
# define a function to convert the string to a dictionary
import ast 
def str_to_dict(s):
    return ast.literal_eval(s)

# apply the function to the column
df['modified_aspects'] = df['modified_aspects'].apply(str_to_dict)
```

```python
for index, row in df.iterrows():
    if row['product_type'] == 'dvd':
        if "director's style" in row['modified_aspects']:
            row['modified_aspects']["directors style"] = row['modified_aspects'].pop("director's style")
        
df      
```
